## List Iterations

1. Design a Python function product_list(nums: list) -> float that returns the product of all numbers in the list (1 for empty list)

```python
def product_list(nums: list) -> float:
    """Returns the product of all the numbers in the list"""
    total = 1
    for n in nums:
        total = n * total
    return total
```

```python
import piplite
await piplite.install(['pytest', 'ipytest'])

import ipytest
ipytest.autoconfig()
```

```python
%%ipytest -qq

def test_product_list():
    assert product_list([1,2,3,4]) == 24
    assert product_list([1,3]) == 3
```

2. Design a Python function count_occurrences(items: list, target) -> int that returns how many times target appears in the list.

```python
def count_occurrences(items: list, target) -> int:
    """Returns how many times target appears in the list"""
    count = 0
    for item in items:
        if item == target:
            count += 1
    return count
```

```python
%%ipytest -qq

def test_count_occurrences():
    assert count_occurrences([1,2,2,3,2], 2) == 3
   
```

3. Design a Python function filter_by_prefix(words: list, prefix: str) -> list that returns a list of all words starting with the given prefix. In Python, if you have a string s, you can check if it starts with a prefix using s.startswith(prefix).

```python
empty: list = []
def filter_by_prefix(words: list, prefix: str) -> list:
    for w in words:
        if w not in empty:
            if w.startswith(prefix):
                empty.append(w)
    return empty
```

```python
%%ipytest -qq

def test_filter_by_prefix():
    assert filter_by_prefix(["Hello", "Hey", "Hi"], "He") == ["Hello", "Hey"]
```

3. Write a function that takes a dictionary of sessions and returns a list of all unique topics covered.

```python
sessions = {
    "CS101": {
      "title": "Introduction to AI",
      "speaker": "Dr. Martinez",
      "topics": ["AI", "Machine Learning"]
    },
    "CS102": {
      "title": "Deep Learning Techniques",
      "speaker": "Prof. Nguyen",
      "topics": ["Neural Networks", "Deep Learning"]
      },
    "CS103": {
      "title": "Quantum Computing Basics",
      "speaker": "Dr. Patel",
      "topics": ["Quantum", "Computing"]
    },
    "CS104": {
      "title": "Cybersecurity Trends",
      "speaker": "Ms. Lee",
      "topics": ["Security", "Networking"]
    }
}

unique_topics: list = []
def get_unique_topics(sessions) -> list:
    for details in sessions.values():
        for topic in details["topics"]:
            if topic not in unique_topics:
                unique_topics.append(topic)
    return unique_topics

print(get_unique_topics(sessions))
```
