## Function Syntax Practice

```python
%%ipytest -qq

def greet(name: str) -> str:
    """produces Hello, name!"""
    return "Hello " + name + "!"

def test_greet():
    assert greet("Rithvik") == "Hello Rithvik!"
```

## Conditionals

```python
%%ipytest -qq

def shipping_cost(weight: int) -> int:
  """if weight <= 1, cost is £5; if weight <= 5, cost is £10; otherwise, £20"""
  if weight <= 1:
    return 5
  elif weight <= 5:
    return 10
  else:
    return 20

def test_shipping_cost():
    assert shipping_cost(1) == 5
    assert shipping_cost(3) == 10
    assert shipping_cost(6) == 20
```

```python
%%ipytest -qq

def grade_letter(score: int) -> str:
    """returns "A" for scores 90 and above, "B" for 80–89, "C" for 70–79, "D" for 60–69, and "F" otherwise"""
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

def test_grade_letter():
    assert grade_letter(92) == "A"
    assert grade_letter(81) == "B"
    assert grade_letter(72) == "C"
    assert grade_letter(2) == "F"
```

## Dataclass

```python
from dataclasses import dataclass

@dataclass
class Book:
    title: str
    author: str
    pages: int

book_1: Book = Book("harry potter", "JK Rowling", 500)
book_2: Book = Book("percy jackson", "Rick Riordan", 450)
book_3: Book = Book("rithvik", "Rithvik", 200)

books: [book_1, book_2, book_3]

big_books = list(filter(lambda b: b.pages > 300, books))

print(big_books)
```
