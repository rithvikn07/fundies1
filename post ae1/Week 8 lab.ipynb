## Problem 1

```python
def book_cost(num_books: int, hardcover: bool) -> int:
  """each paperback costs £12, while hardcover costs £25"""
  if hardcover:
    return num_books * 25
  else:
    return num_books * 12
```

```python
import piplite
await piplite.install(["pytest", "ipytest"])

import ipytest
ipytest.autoconfig()
```

```python
from dataclasses import dataclass
from datetime import date
```

```python
%%ipytest -qq

def test_book_cost():
    assert book_cost(2, True) == 50
    assert book_cost(2, False) == 24
```

```python
def letter_grade(score: int) -> str:
    """assigns letter grade based on numeric score"""
    if 90 <= score <= 100:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"
```

```python
%%ipytest -qq

def test_letter_grade():
    assert letter_grade(91) == "A"
    assert letter_grade(80) == "B"
    assert letter_grade(75) == "C"
    assert letter_grade(50) == "F"
```

## Problem 3

```python
@dataclass
class ToDoItem:
    descr: str
    due: date
    tags: list[str]
```

```python
my_todos = [
    ToDoItem("buy milk", date(2020, 7, 27), ["shopping", "home"]),
    ToDoItem("grade hwk", date(2020, 7, 27), ["teaching"]),
    ToDoItem("meet students", date(2020, 7, 26), ["research"]),
    ToDoItem("buy tickets", date(2020, 7, 30), ["vacation"]),
    ToDoItem("future task", date(2027, 7, 26), ["timetravel"]),
]
```

```python
teachings = list(filter(lambda t: "teaching" in t.tags, my_todos)) 
```

```python
print(teachings)
```

```python
after_2025 = list(filter(lambda d: d.due > date(2025, 1, 1), my_todos))
```

```python
print(after_2025)
```
